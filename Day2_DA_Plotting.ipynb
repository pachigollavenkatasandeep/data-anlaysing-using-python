```python
#An India vs Australia match is on live .India has given 298 as target for Australia,the runs for every 5 overs are gives as follows:
#ind=[25,51,84,131,160,189,220,250,267,297]
#Aus has completed 25 overs
#aus=[15,41,94,110,151].plot overs vs runs for both the countries with India has blue line & aus has Yellow ,line.Give appropriate labels and legend.The marker for Ind is flag and aus is +.
#Take overs from a range function.Create runs as arrays
import numpy as np
ind=np.array([25,51,84,131,160,189,220,250,267,297])
aus=np.array([15,41,94,110,151])
overs=np.arange(5,51,5)
overs1=np.arange(5,26,5)
plt.plot(overs,ind,label='INDIA',color='blue',marker='>',linewidth=2)
plt.plot(overs1,aus,label='Australia',color='yellow',marker='P',linewidth=2)
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.title("INDIA vs AUSTRALIA")
plt.legend()
plt.grid()
```
