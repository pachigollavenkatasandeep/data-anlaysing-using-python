Extracting HTML code of a website
Program flow
1.Import libraies
2.Save the url
3.Using requests.get(),access the web page
4.Using BeautifulSoup(),access the HTML code


```python
from bs4 import BeautifulSoup
import requests as req
```

```python
url="https://colab.research.google.com/drive/1iuQyTvaTNwphRJBUmmEWjSaUKu6kYfwg#scrollTo=KgRDwKm8DPmN"
```

```python
page=req.get(url)
```

```python
code=BeautifulSoup(page.text,'html')
code
```

```python
t=code.find('title')
t
```

```python
url1="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
page=req.get(url1)
code1=BeautifulSoup(page.text,'html')
code1
```

```python
t1=code1.find('table')
t1
```

```python
t2=t1.find_all('th')
t2
```

```python
heading=[i.text for i in t2]
heading
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=heading)
print(mydf)
```
