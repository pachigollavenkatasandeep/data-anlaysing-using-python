```python
#printing tuple
tup=(1,2,6,'Apple',6,'Vasavi','CSE',6)
print(tup)
print(len(tup))#length of tuple
print(tup.count(6))#It gives how many times the given value has repeated in the given tuple
print(tup.index(1))#it returns the index of the given element from the tuple of elements
tup1=(35,26,78,12,6,90)
print(sorted(tup1))
tup2=('Chinnu','Santhosh','Surya','Anand','Mohith')
print(sorted(tup2))
```

```python
#minimum and maximum of tuples
tup1=(35,26,78,12,6,90)
print(min(tup1))
print(max(tup1))
```

```python
#concating two tuples
tup=('APPLE',18,69)
tup1=('Banana',69,True)
tup2=tup+tup1
print(tup2)
```

```python
#converting list to tuple
l1=[1,2,"Apple",6]
t=tuple(l1)
print(t)
```

```python
#updating the value in tuple
tup=(1,'Apple',2,'Banana',3,'Guava')
x=list(tup)
x[1]=2
y=tuple(x)
print(y)
```

```python
#deleting the tuple
tup=(1,'Apple',2,'Banana',3,'Guava')
del tup

```

```python
#replicating tuple
tup=(1,2,3)
x=tup*3
print(x)
```

```python
#printing set
a={1,2,2,2,3,3,4,5,5,6,6,6,7}
print(a)
```

```python
#Adding elements
a={1,2,3,4,5,6}
a.add(10)
print(a)
#Add set of elements to another set
a={1,2,3,4,5}
b={6,7,8,9,10}
a.update(b)
print(a)
```

```python
#Remove element from set
a={1,2,3,4,5}
a.remove(4)
print(a)
a.pop()
print(a)
a.clear()
print(a)
```

```python
#union of elements
set1 = {1, 2, 3,4}
set2 = {1,2,3, 4, 5,6,7}
uni=set1.union(set2)
print(uni)
ini=set1.intersection(set2)
print(ini)
diff=set1.difference(set2)
print(diff)
sub=set1.issubset(set2)
print(sub)
super=set1.issuperset(set2)
print(super)
```
