```python
#Hypothesis testing:
#import
import numpy as np
from scipy import stats
```

```python
data1=np.array([28,21,26,29,36])
data2=np.array([21,27,35,12,28])
print(data1.mean())
print(data2.mean())
```

```python
#Perform t-test assuming unequal variances
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
```

```python
#Print the results
print(("t-statistic:",t_stat))
print("P-value",p_value)
```

```python
#Set significance level
alpha=0.05
```

```python
if p_value < alpha:
  print("Reject the null hypothesis.the means are significant")
else:
  print("Fail to reject the null hypothesis.the means are  not significant")
```

```python
import pandas as pd
df=pd.read_csv('/content/Vaccine.csv')
df
```

```python
group_A=df['Efficiency']
group_B=df['Rec_Rate']
t_stat,p_value=stats.ttest_ind(group_A,group_B)
print("t-Val",t_stat)
print("P-val",p_value)
```

```python
alpha=0.05
```

```python
if p_value > alpha:
  print("H0-related")
else:
  print("H1-not related")
```

```python
dfA=df[df['Vaccine']=='A']['Efficiency']
dfB=df[df['Vaccine']=='B']['Efficiency']
```

```python
dfA
```

```python
f_stat,p_value=stats.f_oneway(dfA,dfB)
print("F-statistic:",f_stat)
print("p-value:",p_value)
```
